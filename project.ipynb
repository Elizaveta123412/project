```python
#работа с железными дорогами и поселками разных типов
for i in range(160183):
    if Russian.loc[i,'type'] == 'железнодорожный':
        if Russian.loc[i, 'name'].split()[1] == 'остановочный':
            Russian.loc[i,'type'] = Russian.loc[i,'type'].replace('железнодорожный','железнодорожный остановочный пункт')
        elif Russian.loc[i, 'name'].split()[1] == 'путевой':
            Russian.loc[i,'type'] = Russian.loc[i,'type'].replace('железнодорожный','железнодорожный путевой пост')
        else:
            Russian.loc[i,'type'] = Russian.loc[i,'type'].replace('железнодорожный','железнодорожный' + ' ' + Russian.loc[i, 'name'].split()[1])
    if Russian.loc[i,'type'] == 'железнодорожная':
        Russian.loc[i,'type'] = Russian.loc[i,'type'].replace('железнодорожная','железнодорожная' + ' ' + Russian.loc[i, 'name'].split()[1])
    if Russian.loc[i,'type'] == 'поселок' and ('сельского' in Russian.loc[i, 'name'] or 'городского' in Russian.loc[i, 'name']):
        Russian.loc[i,'type'] = Russian.loc[i,'type'].replace('поселок', 'поселок' + ' ' + Russian.loc[i, 'name'].split()[1] + ' ' + Russian.loc[i, 'name'].split()[2])
    if Russian.loc[i, 'name'] == 'посёлок' and ('сельского' in Russian.loc[i, 'name'] or 'городского' in Russian.loc[i, 'name']):
        Russian.loc[i,'type'] = Russian.loc[i,'type'].replace('поселок', 'поселок' + ' ' + Russian.loc[i, 'name'].split()[1] + ' ' + Russian.loc[i, 'name'].split()[2])
```
